```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load dataset from /content
df = pd.read_csv("/content/Unemployment_in_India.csv")
print("\nDataset Loaded Successfully:\n")
print(df.head())

# 2. Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# 3. Basic info
print("\nDataset Info:\n")
print(df.info())

# 4. Check for missing values
print("\nMissing Values:\n")
print(df.isnull().sum())

# 5. Line plot - Unemployment over time
plt.figure(figsize=(12,5))
sns.lineplot(data=df, x="Date", y="Estimated Unemployment Rate (%)", marker="o")
plt.title("Unemployment Rate Over Time in India")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. Region-wise unemployment bar chart
plt.figure(figsize=(12,6))
sns.barplot(data=df, x="Region", y="Estimated Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.title("Region-wise Unemployment Rate in India")
plt.tight_layout()
plt.show()

# 7. Correlation heatmap (numeric columns only)
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(7,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

print("\nTask 2 Completed Successfully!")
```
